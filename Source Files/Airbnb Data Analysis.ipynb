**Task 4: Data Cleaning**

```python
data.isnull().sum()

# dropping all missing values rows
data.dropna(inplace=True)

# data.fillna()
data.isnull().sum()
```

**Task 5: Data Analysis**

**Univariate Analysis**

```python
# idenfying outliers in price

## filtering data without outliers:
df = data[data['price'] < 1500]

sns.boxplot(data=df, x='price')
```

```python
df = df.replace([np.inf, -np.inf], np.nan)
```

**Feature Engineering**

```python
# First create the 'price per bed' column
df.loc[:, 'price per bed'] = df['price'] / df['beds']
  # Assuming 'price' and 'beds' columns exist

# Then calculate average price per bed
df.groupby(by='neighbourhood_group')['price per bed'].mean()
```

### Distribution of Price:

```python
plt.figure(figsize=(8,5))
sns.histplot(df['price'], bins=50, kde=True)
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Count')
plt.xlim(0, df['price'].quantile(0.95))  # Focus on 95th percentile to exclude extreme outliers
plt.show()
```

### Count Plot of Room Types:

```python
plt.figure(figsize=(7,5))
sns.countplot(data=df, x='room_type', order=df['room_type'].value_counts().index)
plt.title('Count of Listings by Room Type')
plt.show()
```

### Boxplot of Prices by Neighbourhood Group

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='neighbourhood_group', y='price')
plt.title('Price Variation by Neighbourhood Group')
plt.ylim(0, df['price'].quantile(0.95))  # Exclude extreme prices
plt.show()
```

### Listings Count by Neighbourhood Group and Room Type:

### Correlation Heatmap Including calculated_host_listings_count:

```python
corr_extended = df[['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month',
                    'availability_365', 'beds', 'calculated_host_listings_count', 'price per bed']].corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_extended, annot=True, square=True, cmap='coolwarm')
plt.title('Extended Correlation Matrix')
plt.show()
```

### Price vs. Bedrooms and Baths (after cleaning nulls and converting types):

```python
# Convert 'bedrooms' and 'baths' to numeric if not already done
df['bedrooms'] = pd.to_numeric(df['bedrooms'], errors='coerce')
df['baths'] = pd.to_numeric(df['baths'], errors='coerce')

# Plot with log scale on y-axis for price
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='bedrooms', y='price')
plt.yscale('log')
plt.title('Price Distribution by Number of Bedrooms (Log Scale)')
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='baths', y='price')
plt.yscale('log')
plt.title('Price Distribution by Number of Baths (Log Scale)')
plt.show()
```

### Price Per Bed Variation by Room Type and Neighbourhood Group:

```python
plt.figure(figsize=(12, 7))
sns.barplot(data=df, x='neighbourhood_group', y='price per bed', hue='room_type')
plt.title('Average Price Per Bed by Neighbourhood Group and Room Type')
plt.show()
```

### Most Active Neighbourhoods by Listing Volume and Group:

```python
# Get top 15 neighbourhoods by number of listings
top_nbhds = df['neighbourhood'].value_counts().head(15).index
df_top = df[df['neighbourhood'].isin(top_nbhds)]

plt.figure(figsize=(12, 7))
sns.countplot(data=df_top, 
              y='neighbourhood', 
              hue='neighbourhood_group', 
              order=top_nbhds)
plt.title('Top 15 Neighbourhoods by Listing Count and Group')
plt.xlabel('Number of Listings')
plt.ylabel('Neighbourhood')
plt.legend(title='Neighbourhood Group')
plt.tight_layout()
plt.show()
```

### Price Distribution by Room Type and Neighbourhood Group (Violin Plot):

```python
plt.figure(figsize=(12, 12))
sns.violinplot(data=df, x='neighbourhood_group', y='price', hue='room_type')
plt.ylim(0, df['price'].quantile(0.95))  # Limit y-axis to exclude extreme outliers
plt.title('Price Distribution by Room Type and Neighbourhood Group')
plt.show()
```

### Reviews vs. Price Colored by Room Type (Scatterplot with Regression):

```python
plt.figure(figsize=(10, 7))
sns.scatterplot(data=df, x='price', y='number_of_reviews', hue='room_type', alpha=0.6)
sns.regplot(data=df, x='price', y='number_of_reviews', scatter=False, color='gray')
plt.xlim(0, df['price'].quantile(0.95))
plt.title('Reviews vs Price by Room Type')
plt.show()
```

### Availability Distribution by Neighbourhood Group (Histogram):

```python
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='availability_365', hue='neighbourhood_group', multiple='stack',bins=np.arange(0,400,30))
plt.title('Availability Distribution by Neighbourhood Group')
plt.xticks(np.arange(0,450,30))
plt.show()
```

### Distribution of Minimum Nights by Neighbourhood Group (Boxplot):

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='minimum_nights', hue='neighbourhood_group', bins=30, multiple='stack', kde=False)
plt.title('Minimum Nights Histogram by Neighbourhood Group')
plt.xlim(0, df['minimum_nights'].quantile(0.99))
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.stripplot(data=df, x='neighbourhood_group', y='minimum_nights', size=2, jitter=0.25)
plt.title('Minimum Nights Distribution by Neighbourhood Group (Strip Plot)')
plt.ylim(0, df['minimum_nights'].quantile(0.99))
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='neighbourhood_group', y='minimum_nights')
plt.title('Minimum Nights Distribution by Neighbourhood Group')
plt.show()
```

### Relationship Between Host Listings Count and Price (Scatterplot):

```python
plt.figure(figsize=(10, 7))
sns.scatterplot(data=df, x='calculated_host_listings_count', y='price', alpha=0.5)
plt.yscale('log')
plt.title('Host Listings Count vs Price (Log Scale)')
plt.show()
```

### Grouped Bar Plot of Top Neighbourhoods:

```python
# Get top 10 neighbourhoods by listings count
top_neighs = df['neighbourhood'].value_counts().head(10).index
df_top_bar = df[df['neighbourhood'].isin(top_neighs)]

plt.figure(figsize=(14, 7))
sns.barplot(data=df_top_bar, x='neighbourhood', y='price', hue='room_type', estimator='mean', errorbar=None)
plt.title('Average Price by Room Type for Top 10 Neighbourhoods')
plt.ylabel('Average Price')
plt.xlabel('Neighbourhood')
plt.legend(title='Room Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
